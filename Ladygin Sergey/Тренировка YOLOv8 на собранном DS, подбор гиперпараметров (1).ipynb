```python
from glob import glob

def get_file_paths(folder):

    '''
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ —Ñ–∞–π–ª–æ–≤ –≤ –ø–∞–ø–∫–µ —Å –ø—É—Ç—è–º–∏ –∫ –Ω–∏–º
    Args:
        folder - –∞–¥—Ä–µ—Å –ø–∞–ø–∫–∏
    Return:
        —Å–ø–∏—Å–æ–∫ —Ñ–∞–π–ª–æ–≤ —Å –ø—É—Ç—è–º–∏
    '''

    return glob(folder, recursive=True)
```
